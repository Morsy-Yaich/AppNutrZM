```python
print("Nombre de lignes:", df.shape[0]-1,'repas')
print("Nombre de colonnes:", df.shape[1], '(9 colonnes identifiantes + 67 composantes nutritionnelles)')
```

```python
exclude_columns = ['alim_grp_code','alim_ssgrp_code','alim_ssssgrp_code','alim_grp_nom_fr','alim_ssgrp_nom_fr','alim_ssssgrp_nom_fr','alim_code','alim_nom_fr','alim_nom_sci','','alim_code']
name_columns=['alim_grp_nom_fr','alim_ssgrp_nom_fr','alim_ssssgrp_nom_fr','alim_nom_sci']
```

```python
for col in df.columns:
    if col not in exclude_columns:
        df[col] = df[col].apply(
            lambda x: str(int(float(x[1:].replace(',', '.'))) - 1) if isinstance(x, str) and x.startswith('<') and float(x[1:].replace(',', '.')).is_integer() else
                      str(float(x[1:].replace(',', '.')) - 0.1) if isinstance(x, str) and x.startswith('<') and not float(x[1:].replace(',', '.')).is_integer() else
                      x.replace(',', '.') if isinstance(x, str) else x
        )
```

```python
for col in df.columns:
    if col not in exclude_columns:
        if df[col].isnull().any() or (df[col] == '-').any():
            df[col].replace([np.nan, '-'], 0.001, inplace=True)
```

```python
for col in name_columns:
    if df[col].isnull().any() or (df[col] == '-').any():
        df[col].replace([np.nan, '-'], 999, inplace=True)
```

```python
pd.set_option('display.max_rows', None)
print(df.isnull().sum())
pd.reset_option('display.max_rows')
```

```python
for col in df.columns:
    if col not in exclude_columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
```
